```python
import pandas as pd
import numpy as np
import numpy.ma as ma
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
from pyproj import Geod
from scipy.interpolate import griddata
from tqdm import tqdm
import calendar

# --------------------------------------
# 0. User selection
# --------------------------------------
ssp_target = 'ssp2'

ssp_map = {
    'ssp2': 'rsds_ssp2',
    'ssp3': 'rsds_ssp3',
    'ssp5': 'rsds_ssp5'
}
rsds_col = ssp_map[ssp_target]

# Define efficiency and panel_ratio (these were missing in the original code,
# but essential for the PV potential calculation)
efficiency = 0.15 # Example value: 15% efficiency for solar panels
panel_ratio = 1.0 # Example value: assuming entire cell area can be covered by panels

# --------------------------------------
# 1. Load CMIP6 data
# --------------------------------------
df = pd.read_csv(r"C:\Users\User\Downloads\cmip6_rsds_cnrmfull.csv")
df.columns = df.columns.str.strip()
df["time"] = pd.to_datetime(df["time"])

# Filter only required columns
df = df[["time", "lon", "lat", rsds_col]].rename(columns={rsds_col: "rsds"})

# --------------------------------------
# 2. Generate interpolation grid
# --------------------------------------
num_points = 200
lon_grid = np.linspace(119, 123, num_points)
lat_grid = np.linspace(21, 26, num_points)
lon2d, lat2d = np.meshgrid(lon_grid, lat_grid)

# Create a GeoDataFrame for grid points
grid_points = []
for i in range(lon2d.shape[0]):
    for j in range(lon2d.shape[1]):
        grid_points.append(Point(lon2d[i, j], lat2d[i, j]))
gdf_grid = gpd.GeoDataFrame(geometry=grid_points, crs="EPSG:4326")

# --------------------------------------
# 3. Load county shapefile
# --------------------------------------
county_file_path = r"C:\Users\User\Downloads\直轄市、縣(市)界線1140318\COUNTY_MOI_1140318.shp"
gdf_county = gpd.read_file(county_file_path)

# Reproject to EPSG:4326 if needed
if gdf_county.crs.to_string() != 'EPSG:4326':
    gdf_county = gdf_county.to_crs(epsg=4326)

# --------------------------------------
# 4. Calculate grid cell areas
# --------------------------------------
geod = Geod(ellps="WGS84")
# Calculate area for each cell in the grid.
# Assuming cell_areas[i,j] corresponds to the area of the cell
# where its top-left corner is (lon2d[i,j], lat2d[i,j]).
# The last row/column will have zero area as they can't form a full cell.
cell_areas = np.zeros(lon2d.shape)
for i in range(len(lat_grid)-1):
    for j in range(len(lon_grid)-1):
        # Vertices of the cell
        lons = [lon2d[i,j], lon2d[i,j+1], lon2d[i+1,j+1], lon2d[i+1,j]]
        lats = [lat2d[i,j], lat2d[i,j+1], lat2d[i+1,j+1], lat2d[i+1,j]]
        poly_area, _ = geod.polygon_area_perimeter(lons, lats)
        cell_areas[i,j] = abs(poly_area)  # area in m²

# Flatten cell areas to match the flattened grid points
flat_cell_areas = cell_areas.flatten()

# --------------------------------------
# Spatial Join Grid Points with Counties
# This creates a mapping of each grid point to its containing county
# --------------------------------------
print("Performing spatial join to map grid points to counties...")
gdf_grid_with_county = gpd.sjoin(gdf_grid, gdf_county[['COUNTYNAME', 'COUNTYENG', 'geometry']], how="left", predicate='within')
# Add original grid indices back to easily link with interpolated data
gdf_grid_with_county['grid_idx'] = range(len(gdf_grid_with_county))

# --------------------------------------
# 5. Loop through all 12 months
# --------------------------------------
years = np.arange(2015, 2100)
all_months_cv_data = {} # Dictionary to store CV results for all months

for month_num in range(1, 13):
    target_month_name = calendar.month_name[month_num]
    print(f"\nProcessing month: {target_month_name}")

    # Filter data for this month across all years
    df_month = df[df["time"].dt.month == month_num].copy()

    # Pre-interpolate for all years in this month
    monthly_yearly_interpolated_rsds = {}
    for year in tqdm(years, desc=f"Interpolating {target_month_name} data"):
        df_year = df_month[df_month["time"].dt.year == year][["lon", "lat", "rsds"]].copy()
        if df_year.empty:
            monthly_yearly_interpolated_rsds[year] = None
            continue

        points = df_year[["lon", "lat"]].values
        values = df_year["rsds"].values
        grid_rsds = griddata(points, values, (lon2d, lat2d), method='cubic')
        monthly_yearly_interpolated_rsds[year] = ma.masked_invalid(grid_rsds)

    county_cv_results = []
    
    # Iterate through each county once
    for idx, county_row in tqdm(gdf_county.iterrows(), total=len(gdf_county), desc=f"{target_month_name} - Calculating CV"):
        county_name = county_row['COUNTYNAME']
        county_eng = county_row['COUNTYENG']

        # Get the grid points that fall within this county
        county_grid_mask = (gdf_grid_with_county['COUNTYNAME'] == county_name)
        # Get the indices of these grid points in the flattened grid arrays
        county_grid_indices = gdf_grid_with_county[county_grid_mask]['grid_idx'].values
        
        yearly_outputs = []

        for year in years:
            masked_rsds = monthly_yearly_interpolated_rsds.get(year)
            if masked_rsds is None:
                continue
            
            # Flatten the masked_rsds array for easy indexing
            flat_masked_rsds = masked_rsds.flatten()

            # Select rsds and areas only for points within the current county
            rsds_in_county = flat_masked_rsds[county_grid_indices]
            areas_in_county = flat_cell_areas[county_grid_indices]

            # Mask out invalid rsds values before calculation
            valid_mask = ~rsds_in_county.mask
            rsds_in_county_valid = rsds_in_county[valid_mask]
            areas_in_county_valid = areas_in_county[valid_mask]

            # Calculate total PV potential for this county in this year
            # rsds in W/m², area in m² -> rsds * area = W
            potential = rsds_in_county_valid * areas_in_county_valid * efficiency * panel_ratio
            total_potential = np.sum(potential)

            total_potential_MW = total_potential / 1e6
            yearly_outputs.append(total_potential_MW)

        # Calculate mean, std, CV for the month across years
        if len(yearly_outputs) > 1:
            mean_output = np.mean(yearly_outputs)
            std_output = np.std(yearly_outputs)
            if mean_output != 0:
                cv = std_output / mean_output
            else:
                cv = np.nan
        else:
            mean_output = std_output = cv = np.nan  # Not enough data

        county_cv_results.append({
            "County": county_name,
            "County_Eng": county_eng,
            "Mean_PVOUT_MW": mean_output,
            "Std_PVOUT_MW": std_output,
            "CV": cv
        })

    # Convert to DataFrame
    df_cv = pd.DataFrame(county_cv_results)
    df_cv_sorted = df_cv.sort_values(by="CV", ascending=False)
    df_cv_sorted = df_cv_sorted[df_cv_sorted["CV"].notna()]
    
    all_months_cv_data[month_num] = df_cv_sorted

    # --------------------------------------
    # 6. Plot bar chart for this month
    # --------------------------------------
    plt.figure(figsize=(12,7))
    bars = plt.bar(df_cv_sorted["County_Eng"], df_cv_sorted["CV"])
    
    # Add CV values on top of the bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, round(yval, 3), ha='center', va='bottom', fontsize=9) # Adjusted y-offset
    
    plt.xticks(rotation=60)
    plt.ylabel("Coefficient of Variation (CV)")
    plt.title(f"Interannual {target_month_name} PVOUT Volatility (CV) by County (2015-2099) [{ssp_target}]")
    plt.tight_layout()
    plt.show()

print("\nAnalysis complete. All 12 monthly plots generated with CV values on bars. 📊")
```

```python
# ======================================
# Summary analysis of volatility across all months: Top 3 stable/unstable
# ======================================

# Combine all months' CV data into one DataFrame
summary_list = []

for month_num, df_cv in all_months_cv_data.items():
    for _, row in df_cv.iterrows():
        summary_list.append({
            "Month_Num": month_num,
            "Month_Name": calendar.month_name[month_num],
            "County": row["County"],
            "County_Eng": row["County_Eng"],
            "CV": row["CV"]
        })

df_summary = pd.DataFrame(summary_list)

# --------------------------------------
# Calculate average CV per county
# --------------------------------------
county_cv_mean = df_summary.groupby(["County", "County_Eng"])["CV"].mean().reset_index()

# Sort to find top 3 most unstable (highest CV) and top 3 most stable (lowest CV)
top3_unstable = county_cv_mean.sort_values(by="CV", ascending=False).head(3)
top3_stable = county_cv_mean.sort_values(by="CV", ascending=True).head(3)

print("\n=== Top 3 Most Unstable Counties (Highest Average CV) ===")
print(top3_unstable[["County_Eng", "CV"]].to_string(index=False, formatters={"CV": "{:.3f}".format}))

print("\n=== Top 3 Most Stable Counties (Lowest Average CV) ===")
print(top3_stable[["County_Eng", "CV"]].to_string(index=False, formatters={"CV": "{:.3f}".format}))

# --------------------------------------
# 2. Find the month giving highest and lowest CV for each county
# --------------------------------------
highest_cv_months = df_summary.loc[df_summary.groupby("County")["CV"].idxmax()][["County_Eng", "Month_Name", "CV"]]
lowest_cv_months = df_summary.loc[df_summary.groupby("County")["CV"].idxmin()][["County_Eng", "Month_Name", "CV"]]

print("\n=== Month with Highest CV per County ===")
print(highest_cv_months.to_string(index=False, formatters={"CV": "{:.3f}".format}))

print("\n=== Month with Lowest CV per County ===")
print(lowest_cv_months.to_string(index=False, formatters={"CV": "{:.3f}".format}))
```
